## Prepare Data for Prediction

### Subtask:
Prepare the `df_delhi` DataFrame for building a prediction model by creating lag features for PM2.5 and other relevant pollutants, defining the target variable, and then splitting the data into training and testing sets, ensuring proper time-series split.


**Reasoning**:
To prepare the data for prediction, I will first set the 'Date' column as the index to enable time-series operations. Then, I will generate lag features for various pollutant columns for the past 1, 2, and 3 days, and create the 'PM2.5_target' variable by shifting 'PM2.5' by -1. Afterward, I will remove any rows containing NaN values introduced by these operations, separate the features (X) and target (y), and finally split the dataset into training and testing sets based on an 80/20 time-series split.



````python
print("Preparing df_delhi for prediction model...")


df_delhi.set_index('Date', inplace=True)
print("Date column set as index.")


pollutant_cols = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
for col in pollutant_cols:
    for i in range(1, 4):
        df_delhi[f'{col}_lag{i}'] = df_delhi[col].shift(i)
print("Lag features created.")

df_delhi['PM2.5_target'] = df_delhi['PM2.5'].shift(-1)
print("PM2.5_target created.")

df_delhi.dropna(inplace=True)
print(f"Rows with NaN values dropped. New shape: {df_delhi.shape}")

X = df_delhi.drop(['PM2.5_target'], axis=1)
y = df_delhi['PM2.5_target']
print(f"Features (X) shape: {X.shape}, Target (y) shape: {y.shape}")


split_point = int(len(df_delhi) * 0.8)
X_train, X_test = X.iloc[:split_point], X.iloc[split_point:]
y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]

print(f"Data split into training and testing sets. Training samples: {len(X_train)}, Testing samples: {len(X_test)}")
print("Data preparation complete. First 5 rows of X_train:")
print(X_train.head())
````

## Train Prediction Model

### Subtask:
Train a suitable regression model (e.g., RandomForestRegressor, or a simple linear model) on the prepared training data to predict PM2.5 values. We will use a model that can handle time-series data or leverage the lag features effectively.


**Reasoning**:
To train the prediction model as instructed, I will import `RandomForestRegressor`, instantiate it with the specified parameters, and then fit it to the training data (`X_train` and `y_train`).



````python
from sklearn.ensemble import RandomForestRegressor

print("Training RandomForestRegressor model...")

# Drop the 'City' and 'AQI_Bucket' columns as they are categorical
# and cause a ValueError in the RandomForestRegressor which expects numerical input.
X_train_numeric = X_train.drop(['City', 'AQI_Bucket'], axis=1)
X_test_numeric = X_test.drop(['City', 'AQI_Bucket'], axis=1)

# Instantiate the RandomForestRegressor model
# n_estimators is set to 100 as a suitable number, and random_state for reproducibility
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model using the numerical training data
model.fit(X_train_numeric, y_train)

print("Model training complete.")
````
